```python
#dropping the time, confirnmed indian national and foreign national

covid.drop(['Sno', 'Time', 'ConfirmedIndianNational', 'ConfirmedForeignNational'], inplace = True, axis = 1)
```

```python
covid.head()
```

```python
#changing format of date column

covid['Date'] = pd.to_datetime(covid['Date'], format = '%Y-%m-%d')
```

```python
covid.head()
```

```python
#number of Active Cases

covid['Active Cases'] = covid['Confirmed'] - (covid['Cured'] + covid['Deaths'])
```

```python
covid.head()
```

```python
covid.tail()
```

```python
#Creating a pivot table

statewise = pd.pivot_table(covid, values = ['Cured', 'Deaths', 'Confirmed'], 
                           index = 'State/UnionTerritory', aggfunc = max)
```

```python
#Recovery rate

statewise['Recovery Rate'] = statewise['Cured']*100/statewise['Confirmed']
```

```python
#Mortality rate

statewise['Mortality Rate'] = statewise['Deaths']*100/statewise['Confirmed']
```

```python
#sort the confirmed cases in descending order

statewise = statewise.sort_values(by = ['Confirmed'], ascending = False)
```

```python
#plotting the pivot table using a nice visual

statewise.style.background_gradient(cmap = 'cubehelix')
```

```python
#Remove rows where state is India

vaccine_df = vaccine[vaccine.State != 'India']

vaccine_df
```

```python
#Reanming column

vaccine_df.rename(columns = {'Total Individuals Vaccinated':'Total'}, inplace = True)

vaccine_df
```

```python
#Most Vaccinnated State

max_vac = vaccine_df.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total', ascending = False)[:10]
max_vac
```

```python
#Top 10 states base on number of vaccination

fig = plt.figure(figsize=(16,9))
plt.title('Top 10 states with most vaccination in India', size = 25)
ax = sns.barplot(data =max_vac.iloc[:10], y = max_vac.Total, x = max_vac.index, linewidth = 2, edgecolor = 'red' )

plt.xlabel('States')
plt.ylabel('Vaccination')
plt.show()
```

```python
#least vaccinated states

least_vac = vaccine_df.groupby('State')['Total'].sum().to_frame('Total')
least_vac = least_vac.sort_values('Total', ascending = True) [:5]

least_vac
```

```python
fig = plt.figure(figsize=(20,12))
plt.title('Top 5 states with least vaccination in India', size = 25)
ax = sns.barplot(data = least_vac.iloc[5:], y = least_vac.Total, x = least_vac.index, linewidth = 2, edgecolor = 'red' )

plt.xlabel('States')
plt.ylabel('Vaccination')
plt.show()
```
